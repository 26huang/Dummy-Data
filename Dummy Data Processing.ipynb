```python
import pandas as pd
import datetime
import random
import names
```

```python
def random_date(start, end):
    random_seconds = random.randint(0, int((end - start).total_seconds()))
    return start + datetime.timedelta(seconds=random_seconds)
```

```python
def generate_from(start, end, interval=3600):
    interval_num = int((end-start).total_seconds()/interval)
    return pd.DataFrame({'From Time': [start+datetime.timedelta(seconds=i*interval) for i in range(interval_num)]})
def generate_to(start, end, interval=3600):
    interval_num = int((end-start).total_seconds()/interval)
    return pd.DataFrame({'To Time': [start+datetime.timedelta(seconds=(i+1)*interval) for i in range(interval_num)]})
```

```python
start_input = "1/1/2008 12:00 AM"
end_input = "1/2/2008 12:00 AM"
format = '%m/%d/%Y %I:%M %p'
```

```python
start = datetime.datetime.strptime(start_input, format)
end = datetime.datetime.strptime(end_input, format)
```

```python
def generate_student(start, end, n):
    name = names.get_full_name()
    logins = [random_date(start, end) for i in range(n)]
    return pd.DataFrame({name: logins})
```

```python
def generate_class(start, end, n=10):
    df = pd.concat([generate_student(start, end, random.randint(0, n)) for i in range(n)], axis=1)
    return df
```

```python
data = generate_class(start, end)
```

```python
data = data.fillna(False)
```

```python
data.to_excel('dummy_data.xlsx')
```

```python
from_time = generate_from(start, end)
to_time = generate_to(start, end)
```

```python
time_data = pd.concat([from_time, to_time], axis=1)
```

```python
time_data.to_excel('to_from_time.xlsx')
```

```python
processed_data = pd.DataFrame()
for i in data.columns:
    processed_data[i] = data[i].apply(lambda x: x.replace(minute=0, second=0) if x != False else False)
```

```python
processed_data
```

```python
for i in processed_data.columns:
    time_data[i] = time_data['From Time'].apply(lambda x: 1 if x in list(processed_data[i]) else 0)
```

```python
time_data.to_excel('dummy_data.xlsx')
```
